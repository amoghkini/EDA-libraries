```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
data = sns.load_dataset('flights')
data.head()
```

### Create crosstab/pivot table first

```python
x = data.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
x
```

```python
sns.heatmap(x)
```

## Darker shade shows lower value and lighter shade shows higher value

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x,linewidths=0.8)
```

```python
sns.heatmap(x,linewidths=0.8,annot=True,fmt='d')
```

```python
sns.heatmap(x,linewidths=0.8,annot=True,fmt='d',cmap='viridis')
```

```python
sns.heatmap(x,linewidths=0.8,annot=True,fmt='d',cmap='summer')
```

```python
sns.heatmap(x,linewidths=0.8,annot=True,fmt='d',cmap='winter')
```
