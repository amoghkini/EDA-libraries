```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
data = sns.load_dataset('tips')
data.head()
```

The advantage of the iolin plot oer the box plot is that aside from showing the above mentioned statistics it also shows the entire distribution of the data

We can say that it is combination of boxplot and Probability/Kernal Density Function

They are often used to compare the distribution of a fiven ariable across some categories

```python
sns.violinplot(data.total_bill)
```

```python
sns.violinplot(x='day',y='total_bill',data=data,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=data,hue='sex',split=True)
```
