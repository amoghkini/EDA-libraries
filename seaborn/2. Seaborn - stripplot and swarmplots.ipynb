```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
data = sns.load_dataset('tips')
data.head()
```

## 1. Stripplot

```python
sns.catplot(x='day',y='tip',data=data,kind='strip')
```

### We internally use jitter parameter to seperate the points of same values from each other. It creates noice and makes charts interesting.

```python
sns.catplot(x='day',y='tip',data=data,jitter=1.2,kind='strip')
```

```python
sns.catplot(x='day',y='tip',data=data,jitter=0,kind='strip')
```

```python
sns.stripplot(x='day',y='tip',data=data,jitter=0)
```

## 2. Swarmplot

```python
sns.catplot(x='day',y='tip',kind='swarm',data=data)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=data,hue='sex')
```

```python
sns.swarmplot(x='day',y='tip',data=data,hue='sex')
```
