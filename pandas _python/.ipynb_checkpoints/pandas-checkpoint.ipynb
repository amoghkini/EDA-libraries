#### 1. Find top 5 batsmen in terms of run score

```python
## we will apply group by function on batsmen
```

```python
batsmen_score = deliveries.groupby('batsman')
```

```python
batsmen_score.size()
```

```python
batsmen_score.get_group('V Kohli').shape
```

```python
batsmen_score.get_group('V Kohli').over.count()
```

```python
batsmen_score.get_group('V Kohli').batsman_runs.sum()
```

```python
batsmen_score.batsman_runs.sum().sort_values(ascending=False).head()
```

#### 2. Find out batsmen with most no of 6's.

```python
mask = deliveries['batsman_runs'] == 6
delivery_final = deliveries[mask]
```

```python
delivery_final.batsman_runs.count()
```

```python
group_bats  = delivery_final.groupby('batsman')
```

```python
group_bats.size().sort_values(ascending=False).head()
```

#### 3. Virat's most no of runs vs teams

```python
vk = deliveries[deliveries['batsman'] == 'V Kohli']

ans = vk.groupby('bowling_team')

ans.batsman_runs.sum().sort_values(ascending=False).head()
```

```python
def func_most_runs_against_team(name):
    vk = deliveries[deliveries['batsman'] == name]

    ans = vk.groupby('bowling_team')

    return ans.batsman_runs.sum().sort_values(ascending=False).head(1)
    
```

```python
func_most_runs_against_team('V Kohli')
```

```python
func_most_runs_against_team('DA Warner')
```

## isin() function

### Find the most destructuive death over batsman in the history of IPL
### logic:
### strike_rate = (no of runs/no of balls)/100
### min batsman 200 balls in over 16-20

```python
death_oves = deliveries[deliveries.over > 15]
```

```python
all_bats = death_oves.groupby('batsman')['batsman_runs'].count() 
```

```python
## finds players who played more than 200 balls
index_bats = all_bats[all_bats > 200].index
```

```python
type(index_bats)
```

```python
index_bats = all_bats[all_bats > 200].index.tolist()
```

```python
type(index_bats)
```

```python
## find runs scored and balls played by these 43 batsmen
```

```python
final = deliveries[deliveries.batsman.isin(index_bats)]
```

```python
(final.groupby('batsman').batsman_runs.sum() / final.groupby('batsman').batsman_runs.count() * 100).sort_values(ascending=False)
```

## Merge

### Find orange cap holder

```python
match = pd.read_csv('matches - matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
merge_df = delivery.merge(match,left_on='match_id',right_on='id')
```

```python
merge_df.groupby(['season','batsman']).batsman_runs.sum()
```

```python
### we got season and batsman as index. We will reset the index
```

```python
merge_df.groupby(['season','batsman']).batsman_runs.sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep="first").sort_values(by='season')[['season','batsman']].reset_index()
```
