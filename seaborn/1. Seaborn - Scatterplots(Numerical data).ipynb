# Relational/Statistical Plots
Relational plots are used to perform multivariate analysis

1. Scatter Plots
2. Lines Plots (Specifically used for time series data)
3. FacetGrids (Plotting multiple graphs side by side)

## 1. Scatter plots

It can be created using two ways
1. relplot [Figure Level Function]
2. scatterplot [Axes level function]

We use three aditional variables <b>hue,size,style</b>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data.head()
```

## Using relplot() method

### Plot a scatter plot between total bill and tips(numerical data)

```python
sns.relplot(x='total_bill',y='tip',data=data)
```

```python
## Pass kind parameter
sns.relplot(x='total_bill',y='tip',data=data,kind='scatter')
```

### hue parameter

```python
sns.relplot(x='total_bill',y='tip',data=data,kind='scatter',hue='smoker')
```

### The style parameter

```python
sns.relplot(x='total_bill',y='tip',data=data,kind='scatter',hue='smoker',style='sex')
```

### The size parameter

In hue and style parameter, we pass categorical data <br>
In size parameter we pass continuous or numberical data

```python
sns.relplot(x='total_bill',y='tip',data=data,kind='scatter',hue='smoker',style='sex',size='size',sizes = (15,100))
```

## Using scatterplot method

```python
sns.scatterplot(x='total_bill',y='tip',data=data,hue='smoker',style='sex',size='size',sizes = (15,100))
```
