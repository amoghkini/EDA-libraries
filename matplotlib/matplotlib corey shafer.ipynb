```python
import csv
from collections import Counter
f = open('data.csv')
csv_reader = csv.DictReader(f)
#row = next(csv_reader)
language_ccounter = Counter()
for row in csv_reader:
    language_ccounter.update(row["LanguagesWorkedWith"].split(';'))
#print(language_ccounter)

language = []
popularity = []

for item in language_ccounter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
    
plt.barh(language[::-1],popularity[::-1])
plt.title("Most Popular Languages")
plt.ylabel("Programming Language")
plt.xlabel('Number of people who use')

plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
age = df['Age']
dev_salaries = df.All_Devs
py_salaries = df.Python
js_salaries = df.JavaScript

plt.plot(age,dev_salaries,linestyle='--',linewidth = 0.9,label="Al devs")

plt.plot(age,py_salaries,color='#444444',linewidth=1.2,label='Python')

plt.fill_between(age,py_salaries,alpha=0.25)

plt.legend()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
age = df['Age']
dev_salaries = df.All_Devs
py_salaries = df.Python
js_salaries = df.JavaScript

plt.plot(age,dev_salaries,linestyle='--',linewidth = 0.9,label="All devs")

plt.plot(age,py_salaries,color='#444444',linewidth=1.2,label='Python')
median = 57287

plt.fill_between(age,py_salaries,median,where=(py_salaries > median),interpolate=True,alpha=0.25)
plt.fill_between(age,py_salaries,median,where=(py_salaries <= median),interpolate=True,color='red',alpha=0.25)

plt.legend()
plt.show()
```
